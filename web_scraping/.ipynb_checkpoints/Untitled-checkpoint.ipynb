```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(p12.text,'lxml')
```

```python
type(s)
```

```python
print(s.prettify())
```

```python
secciones = s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
secciones
```

```python
seccion = secciones[0]
```

```python
seccion.a.get_text()
```

```python
seccion.a.get('href')
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
```

```python
links_secciones
```

```python
sec = requests.get(links_secciones[0])
```

```python
sec.status_code
```

```python
s_seccion = BeautifulSoup(sec.text, 'lxml')
```

```python
print(s_seccion.prettify())
```

```python
featured_article = s_seccion.find('div', attrs={'class':'article-item__header deco-bar-here'})
featured_article
```

```python
featured_article.a.get('href')
```

```python
article_list = s_seccion.find('div', attrs={'class':'article-item__content'}).find_all("li")
```

```python
article = article_list[0]
```

```python
article_title = [article.a.get_text() for article in article_list]
article_title
```
