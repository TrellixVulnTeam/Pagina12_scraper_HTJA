```python
def obtener_notas(soup):
    
    lista_notas = []
    
    featured_article = soup.find('div', attr={'class':'article-item__header deco-bar-here'})
    if featured_article:
        lista_notas.append(featured_article.a.get('href'))
        
    article_list = soup.find('div', attrs={'class':'articles-list'})
    for article in article_list.find_all('href'):
        if article.a:
            lista_notas.append(article.a.get('href'))
            
    return lista_notas    
```

```python
lista_notas = obtener_notas(s_seccion)
lista_notas
```
